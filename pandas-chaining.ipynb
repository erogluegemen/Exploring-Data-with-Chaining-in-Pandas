```python
# a glorious function
def tweak_autos(autos:pd.core.frame.DataFrame):
    cols = ['city08', 'comb08', 'highway08', 'cylinders', 'displ', 'drive', 'eng_dscr', 'fuelCost08', 'make', 'model', 'trany', 'range', 'createdOn', 'year']
    return (autos
            [cols]
            .assign(cylinders=autos.cylinders.fillna(0).astype('int8'),
                    displ=autos.displ.fillna(0).astype('float16'),
                    drive=autos.drive.fillna('Other').astype('category'),
                    automatic=autos.trany.str.contains('Auto'),
                    speeds=autos.trany.str.extract(r'(\d)+').fillna('20').astype('int8'),
                    createdOn=pd.to_datetime(autos.createdOn.replace({' EDT': '-4:00', ' EST': '-05:00'}, regex=True)),
                    ffs=autos.eng_dscr.str.contains('FFS')
                   )
           .astype({'highway08':'int8', 'city08':'int16', 'comb08':'int16', 
                    'fuelCost08':'int16', 'range':'int16', 'year':'int16', 'make':'category'})
           .drop(columns=['trany', 'eng_dscr'])
          )

tweak_autos(autos)    
```

```python
# Easy to debug
# - assign to var (df3)
# - comment out
# - pipe to display

from IPython.display import display

def get_var(df:pd.core.frame.DataFrame, var_name:str) -> pd.core.frame.DataFrame:
    globals()[var_name] = df
    return df

def tweak_autos(autos:pd.core.frame.DataFrame):
        return (autos
            [cols]
            # create var
            .pipe(get_var, 'df3')
            .assign(cylinders=autos.cylinders.fillna(0).astype('int8'),
                    displ=autos.displ.fillna(0).astype('float16'),
                    drive=autos.drive.fillna('Other').astype('category'),
                    automatic=autos.trany.str.contains('Auto'),
                    speeds=autos.trany.str.extract(r'(\d)+').fillna('20').astype('int8'),
                    createdOn=pd.to_datetime(autos.createdOn.replace({' EDT': '-4:00', ' EST': '-05:00'}, regex=True)),
                    ffs=autos.eng_dscr.str.contains('FFS')
                   )
           # debug pipe
           .pipe(lambda df: display(df) or df)
           .astype({'highway08':'int8', 'city08':'int16', 'comb08':'int16', 
                    'fuelCost08':'int16', 'range':'int16', 'year':'int16', 'make':'category'})
           .drop(columns=['trany', 'eng_dscr'])
          )
    
autos2 = tweak_autos(autos)
```

```python
(autos2
 .groupby('year')
 [['comb08', 'speeds']]
 .mean(numeric_only=True)
 .plot()
);
```

```python
# add country
(autos2
 .assign(country=autos2.make.apply(change_country))
 .groupby(['year', 'country'])
 .mean(numeric_only=True)
)
```

```python
# can go deeper and apply multiple aggregates
def second_to_last(ser):
    return ser.iloc[-2]

(autos2
 .assign(country=autos2.make.apply(change_country))
 .groupby(['year', 'country'])
 .agg(['min', 'mean', second_to_last])
)
```

```python
# back to simpler example, adding plots
(autos2
 .assign(country=autos2.make.apply(change_country))
 .groupby(['year', 'country'])
 .mean(numeric_only=True)
 .plot()
)
```

```python
(autos2
 .assign(country=autos2.make.apply(change_country))
 .groupby(['year' ,'country'])
 .mean(numeric_only=True)
 .unstack()
)
```

```python
(autos2
 .assign(country=autos2.make.apply(change_country))
 .groupby(['year' ,'country'])
 .mean(numeric_only=True)
 .unstack()
 .city08
 .rolling(2)
 .mean(numeric_only=True)
 .plot()
 .legend(bbox_to_anchor=(1,1))
)
```
